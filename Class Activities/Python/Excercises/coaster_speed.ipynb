```python
# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create the X and Y axis lists
time = np.arange(0,130,10)
danger_drop_speeds = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
railgun_speeds = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
# Plot the charts and apply some styling
danger_drop, = plt.plot(time, danger_drop_speeds, color="purple", label="danger drop")
railgun, = plt.plot(time, railgun_speeds, color="orange", label="Railgun")
```

```python
# Add labels to X and Y axes :: Add title
plt.title("Roller Coaster Speeds")
plt.ylabel("Speed(MPH)")
plt.xlabel("Run Time")
plt.show()

```
